```python
import requests
import json 

xx = requests.get("https://api.scryfall.com/cards/search?q=c%3Awhite+cmc%3D1")

names = []
oracles = []
for x in xx.json()["data"]:
  oracles.append(x["oracle_text"] if "oracle_text" in x else "")
  names.append(x["name"])

with open("names.txt","w+") as file:
  file.writelines([name+"\n" for name in names])

with open("oracles.txt","w+") as file:
  file.writelines([oracle+"\n" for oracle in oracles])
```
